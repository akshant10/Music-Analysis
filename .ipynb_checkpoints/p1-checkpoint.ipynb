```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
#extract information from wikipedia
#get information about top 100
linkk='https://en.wikipedia.org/wiki/Billboard_Year-End_Hot_100_singles_of_'

rate=[]
song=[]
artist=[]
year=[]


for j in range(1980,2022,1): #use for loop to extract
    
    try:
    
        
        
        page = requests.get(linkk+str(j)) #request data for above link

        soup = BeautifulSoup(page.content, 'html.parser')
        #print(soup)
        tds=soup.find_all('td')


        for i in range(0,300,3):
            rate.append(int(tds[i].text))


        for i in range(1,300,3):
            song.append(tds[i].text.replace('"',''))


        for i in range(2,300,3):
            artist.append(tds[i].text.replace("\n",''))
            year.append(j) 
        
        print(j)
    except:
        
        
        page = requests.get(linkk+str(j)) #request data for above link

        soup = BeautifulSoup(page.content, 'html.parser')
        #print(soup)
        tds=soup.find_all('td')


        for i in range(2,302,3):
            rate.append(int(tds[i].text))

        for i in range(3,302,3):
            song.append(tds[i].text.replace('"',''))


        for i in range(4,302,3):
            artist.append(tds[i].text.replace("\n",''))
            year.append(j)
        print(j)
```

```python
data={'rating':rate,'song_name':song, 'artist':artist,'year':year}

df=pd.DataFrame(data)
```
