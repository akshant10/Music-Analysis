# visualization

## songs count (as per dataset)

```python
df1=ds.groupby('release_date').count()['Unnamed: 0'].reset_index()
plt.plot(df1['release_date'],df1['Unnamed: 0'])
```

## Do new songs have higher duration?

```python
df2=ds.groupby('release_date').mean()['len'].reset_index()
plt.plot(df2['release_date'],df2['len'])
```

## Are new songs romantic?

```python
df3=ds.groupby('release_date').mean()['romantic'].reset_index()
plt.plot(df3['release_date'],df3['romantic'])
```
